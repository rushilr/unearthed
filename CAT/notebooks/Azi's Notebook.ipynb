```python
# need to figure out how to clean timestamp in events dataset
def clean_event_date_time(event):
    real_time = event.OCAL[0:8]
    pm_or_am = event.OCAL[-2:]
    
    cleaned_time = real_time + pm_or_am
    cleaned_time = cleaned_time.replace(".", ":")
    
    cleaned_date = event.OCUR_TS_L
    
    cleaned_date_time = cleaned_date + " " + cleaned_time

    return parser.parse(cleaned_date_time)
```

```python
test = logger1.iloc[3]

def check_for_new_logger_pull(log):
    delta_1hr = datetime.timedelta(hours=1)
    
    if log.timestamp + delta_1hr < log.lead_timestamp:
        return True
    else:
        return False
        
```

```python
logger1['is_pull_time'] = logger1.apply(check_for_new_logger_pull, axis=1)
```

```python
logger1 = logger1.reset_index()
```

```python
pull_times = []
i=0

for index, row in logger1.iterrows():
    
    if row.is_pull_time:
        if not pull_times:
            pull_times.append(row.timestamp)
            logger1.ix[index, 'previous_pull'] = datetime.datetime(2000, 1, 1, 0, 0, 0)
            logger1.ix[index, 'session'] = i
        else:
            logger1.ix[index, 'previous_pull'] = pull_times[i]
            logger1.ix[index, 'session'] = i
            i += 1
            pull_times.append(row.timestamp)
    else:
        logger1.ix[index, 'session'] = i
        
```

```python
logger1[logger1.is_pull_time].head()
```

```python
session_events_dict = {}
for index, row in logger1.iterrows():
    if row.is_pull_time:
        time_of_pull = row.timestamp
        previous_pull = row.previous_pull
        delta_1hr = datetime.timedelta(hours=1)
        
        events_in_past_hour = op_events[(op_events.MACH_SER_NO == 'Truck 1') & (op_events.timestamp >= time_of_pull - delta_1hr) \
                                     & (op_events.timestamp < time_of_pull)]
        
        if events_in_past_hour.shape[0] > 0:
            logger1.ix[index, 'is_candidate_positive_example'] = True
            session_events_dict[row.session] = events_in_past_hour
        else: 
            logger1.ix[index, 'is_candidate_positive_example'] = False
            
        events_between_current_and_last_pull = \
        op_events[(op_events.MACH_SER_NO == 'Truck 1') & (op_events.timestamp <= time_of_pull) \
        & (op_events.timestamp >= previous_pull)]

        if events_between_current_and_last_pull.shape[0] == 0:
            logger1.ix[index, 'is_negative_example'] = True
        else: 
            logger1.ix[index, 'is_negative_example'] = False
    else:
        logger1.ix[index, 'is_candidate_positive_example'] = False
        logger1.ix[index, 'is_negative_example'] = False
```

```python
def find_events_for_session(logger_df, session, truck='Truck 1'):
    pull_time_log = logger_df[(logger_df.session == session) & logger_df.is_pull_time]
    
    delta_1hr = datetime.timedelta(hours=1)
    time_of_pull = list(pull_time_log.timestamp)[0]
    
    return op_events[(op_events.MACH_SER_NO == truck) & (op_events.timestamp >= time_of_pull - delta_1hr) \
                                     & (op_events.timestamp < time_of_pull)]
```
